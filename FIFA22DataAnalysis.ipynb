## Data Exploration and Cleaning

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('players_22.csv')
```

```python
l = data[['sofifa_id','player_url','dob','club_jersey_number','body_type','real_face',
          'player_tags','player_traits', 'goalkeeping_diving','goalkeeping_handling',
          'goalkeeping_kicking','goalkeeping_positioning', 'goalkeeping_reflexes','goalkeeping_speed',
          'ls', 'st', 'rs','lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm',
          'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb','rb','gk',
          'player_face_url','club_logo_url','club_flag_url','nation_logo_url','nation_flag_url']]
```

```python
l.head(2)
```

```python
#Removed all the unnecessary columns
data = data.drop(labels=l, axis=1)
```

```python
data.head()
```

```python
data.describe()
```

```python
#Checking Null values
null = data.isnull().sum()
print(null[null>0])
```

```python
data['value_eur'] = data['value_eur'].fillna(data['value_eur'].mean())
data['wage_eur'] = data['wage_eur'].fillna(data['wage_eur'].mean())
data['club_team_id'] = data['club_team_id'].fillna('No ID')
data['club_name'] = data['club_name'].fillna('Free Agent')
data['league_name'] = data['league_name'].fillna('No League')
data['league_level'] = data['league_level'].fillna('No Level')
data['club_position'] = data['club_position'].fillna('No Position')
data['club_loaned_from'] = data['club_loaned_from'].fillna('Not loaned')
data['club_joined'] = data['club_joined'].fillna('Not joined')
data['club_contract_valid_until'] = data['club_contract_valid_until'].fillna('No validity')
data['nation_team_id'] = data['nation_team_id'].fillna('No team id')
data['nation_position'] = data['nation_position'].fillna('No Position')
data['nation_jersey_number'] = data['nation_jersey_number'].fillna('No Number')
data['release_clause_eur'] = data['release_clause_eur'].fillna(0)
data['pace'] = data['pace'].fillna(data['pace'].mean())
data['shooting'] = data['shooting'].fillna(data['shooting'].mean())
data['passing'] = data['passing'].fillna(data['passing'].mean())
data['defending'] = data['defending'].fillna(data['defending'].mean())
data['dribbling'] = data['dribbling'].fillna(data['dribbling'].mean())
data['physic'] = data['physic'].fillna(data['physic'].mean())
```

```python
#Checking Null values
null = data.isnull().sum()
print(null[null>0])
```

## Position Statistics

```python
data['player_positions']
```

```python
football = []
l=[]
for i in data['player_positions']:
    l.append(i.replace(" ",''))
for j in l:
    if 'ST' in j or 'CF' in j or'LW' in j or 'RW' in j:
        football.append("Forward")
    elif 'CAM'in j or 'CDM' in j or 'CM' in j or 'RM' in j or 'LM' in j:
        football.append('Midfield')
    elif 'GK' in j:
        football.append('Goalkeeper')
    elif 'CB' in j or 'RB' in j or 'LB' in j or 'LWB' in j or 'RWB' in j:
        football.append("Defender")
data['Positions'] = football
```

```python
data['Positions']
```

## Chi Square Test

```python
#Chi square test
data.loc[(data['overall']<50,'Overall_Desc')] = 'Poor'
data.loc[(data['overall']>=50) & (data['overall']<=69.9),'Overall_Desc'] = 'Below Average'
data.loc[(data['overall']>=70) & (data['overall']<=79.9),'Overall_Desc'] = 'Average'
data.loc[(data['overall']>=80) & (data['overall']<89.9),'Overall_Desc'] = 'Good'
data.loc[(data['overall']>=90) & (data['overall']<100),'Overall_Desc'] = 'Best'
```

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
col = ['weak_foot','preferred_foot','international_reputation','skill_moves','work_rate','Overall_Desc']
categ = pd.DataFrame(data = data, columns=col)
categ['weak_foot'] = le.fit_transform(categ['weak_foot'])
categ['preferred_foot'] = le.fit_transform(categ['preferred_foot'])
categ['work_rate'] = le.fit_transform(categ['work_rate'])
categ['Overall_Desc'] = le.fit_transform(categ['Overall_Desc'])
X = categ.drop('Overall_Desc',axis = 1)
Y = categ['Overall_Desc']
```

```python
from sklearn.feature_selection import chi2
chi_scores = chi2(X,Y)
chi_scores = pd.Series(chi_scores[0], index = X.columns)
chi_scores.plot(kind = 'barh')
```

## Prediction using Linear Regression

```python
#LinearRegression
data['value_eur'] = data['value_eur'] / 1000000
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X = pearson.drop('height_cm',axis = 1)
y = data['value_eur']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
lr = LinearRegression()
lr.fit(X_train, y_train)
predict = lr.predict(X_test)
print('Slopes :',lr.coef_)
print('Intercept',lr.intercept_)
```

```python
from sklearn import metrics
print("Mean Squared Error:",metrics.mean_squared_error(y_test,predict))
print("Mean Absolute Error:",metrics.mean_absolute_error(y_test,predict))
print("Root Mean Squared Error:",np.sqrt(metrics.mean_squared_error(y_test,predict)))
```
